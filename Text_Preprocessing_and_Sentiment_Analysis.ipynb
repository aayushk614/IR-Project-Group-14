```python
# Define the list of non relevant keywords
keywords = ["JustinTrudeau","Mitch McConnell","usa", "united states", "trump", "america", "democrat", "cia", "fbi", "nsa", "rahul dravid", "barney frank", "youtube", "White House", "Nancy Pelosi", "US Senate", "Hillary Clinton", "Mahatma Gandhi", "Churchill", "Duncan Hunter", "Uhunmwode"]

# Create an empty list to store the indices of tweets to be removed
remove_indices = []

# Iterate through each tweet in the dataframe
for i in range(len(df)):
    try:
        # Get the full_text column of the tweet
        tweet_text = df.loc[i, "full_text"]
        
        # Check if the tweet contains any of the irrelevant keywords
        for keyword in keywords:
            if keyword in str(tweet_text).lower():
                remove_indices.append(i)
                break
    except:
        pass

# Remove the tweets with irrelevant keywords
df = df.drop(remove_indices)

# Remove tweets from countries other than India
df = df[df["Country"] == "India"]
```

```python
import re

def clean_text(text, contraction_mapping):
    # remove unnecessary spaces
    text = text.strip()
    
    # remove URLs
    text = re.sub(r'http\S+', '', text)
    
    # remove @mentions
    text = re.sub(r'@\w+', '', text)
    
    # replace contractions
    for word in text.split():
        if word.lower() in contraction_mapping:
            text = text.replace(word, contraction_mapping[word.lower()])
    
    # remove non-alphabetic characters and convert to lowercase
    text = re.sub(r'[^a-z# ]', '', text.lower())
    
    # remove extra spaces
    text = re.sub(r'\s+', ' ', text)
    
    return text
```
